**b. Classification**

```python
df_new['CategoricalRating'] = pd.cut(x=df_new['rating'], bins=[18,36.9,55.7,74.5],labels=['Bad','Neutral','Good'])
df_new.head()
```

```python
df_new.drop(["rating"],axis='columns',inplace=True)
df_new.head()
```

```python
plt.figure(figsize=(7,7))
df_new.groupby('CategoricalRating')['CategoricalRating'].count().plot.pie(autopct = '%2.0f')
plt.show()
```

```python
y=df_new.values[:,9]
```

```python
X=df_new.values[:,0:9]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

**SVM**

```python
model_svc = SVC(kernel = 'linear', C = 1E10)
model_svc.fit(X, y)
y_pred = model_svc.predict(X_test)
print("Accuracy:",accuracy_score(y_test, y_pred))
print()
print("Confusion Matrix")
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

**DECISION TREE**

```python
model_dtc = DecisionTreeClassifier()
model_dtc.fit(X_train,y_train)
y_pred = model_dtc.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print()
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix")
print(cm)
```

```python
fig = plt.figure(figsize=(30,30))
_ = tree.plot_tree(model_dtc,class_names=df_new['CategoricalRating'])
```

```python
model_dtcc = DecisionTreeClassifier(criterion="entropy", max_depth=3)
model_dtcc.fit(X_train,y_train)
y_pred = model_dtcc.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print()
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix")
print(cm)
```

```python
text = tree.export_text(model_dtc)
fig = plt.figure(figsize=(30,30))
_ = tree.plot_tree(model_dtc,class_names=df_new['CategoricalRating'])
```

**NAIVE BAYES**

```python
model_nb = GaussianNB()
model_nb.fit(X_train , y_train)
y_predicted = model_nb.predict(X_test)
cm = confusion_matrix(y_test, y_predicted)
print("Accuracy: ",metrics.accuracy_score(y_test, y_predicted))
print()
print("Confusion Matrix")
print(cm)
```

**RANDOM FOREST**

```python
model_rfc = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
model_rfc.fit(X_train, y_train)
y_pred = model_rfc.predict(X_test)
print("Confusion Matrix")
cm = confusion_matrix(y_test, y_pred)
print(cm)
print()
print("Accuracy: ",metrics.accuracy_score(y_test, y_pred))
```

**4. Model Comparison & Validation**

```python
kfold = model_selection.KFold(n_splits=10)
M = []

s = SVC(kernel = 'linear', C = 1E10)
d = DecisionTreeClassifier()
n = GaussianNB()
r = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)

M.append(('SVM', s))
M.append(('DT', d))
M.append(('NB', n))
M.append(('RF', r))

results = []
names = []
scoring = 'accuracy'
for name, model in M:
    kfold = model_selection.KFold(n_splits=10)
    cv_results = model_selection.cross_val_score(model, X_train, y_train, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    Mean_SDV = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(Mean_SDV)
```

The first value shows mean accuracy, and the bracket values give standard deviation of the accuracy of each model.

SVM comes out to be our best model with the highest accuracy and smallest standard deviation. On second, we get Nave Bayes Classifier of 82% accuracy with 13.9 standard deviation. Also note that, the Random Forest model (an ensemble of decision trees) gives a better accuracy mean than a single decision tree!

```python
fig = plt.figure(figsize=(12,12))
fig.suptitle('Models Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
R
```
